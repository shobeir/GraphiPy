```python
from usde.usde import USDE

# Youtube Api's credentials
api_credentials = {'api_key': ' ',	
           'client_secret': "client_secrets.json"}

# Create usde
usde = USDE()
# Create YouTube
youtube = usde.get_youtube(api_credentials)
```

# Searching for videos with topic "USC"

```python
# Create a graph and naming the graph type inside usde first and then modify the graph
graph = usde.create_graph("pandas")
youtube.fetch_videos_by_topic(graph, "usc")
nodes = graph.get_nodes()
df = nodes["video"]
df.head(10)
```

# Visualization with NetworkX

```python
import networkx as nx

nx_graph = nx.Graph();

youtube.fetch_video_by_id_with_comments(info_of_one_video_graph, "DAV2sWfiZN4")
nodes = info_of_one_video_graph.get_nodes()
edges = info_of_one_video_graph.get_edges()
```

```python
print(nodes.keys())
```

```python
print(edges.keys())
```

```python
# create graph from edgelist dataframes
for key in edges:
    
    new_graph = nx.from_pandas_edgelist(
        edges[key], source="Source", target="Target", edge_attr=True)
    
    nx_graph = nx.compose(nx_graph, new_graph)
```

```python
# Add node attributes
for key in nodes:
    df = nodes[key]
    
    for index, row in df.iterrows():
        _id = row["_id"]
        node = nx_graph.node[_id]
        
        for row_key in row.keys():
            node[row_key] = row[row_key]
```

```python
# Draw the graph without node labels
nx.draw(nx_graph)
```

```python
# Draw the graph with node labels
labels = dict((node, data["Label"]) for node, data in nx_graph.nodes(data=True))
nx.draw(nx_graph, labels=labels)
```

```python
pos = nx.spring_layout(nx_graph)
edge_labels = {}
for edge in nx_graph.edges(data=True):
    key = (edge[0], edge[1])
    edge_labels[key] = edge[2]["Label"]
```

```python
# Draw networkx labels
nx.draw_networkx_edge_labels(nx_graph, pos, edge_labels=edge_labels, font_color="red")
```
