# Using GraphiPy to extract data from Youtube

```python
from graphipy.graphipy import GraphiPy
```

```python
# Create graphipy
graphipy = GraphiPy()
```

# Creating the Youtube Object
GraphiPy's Youtube object needs the Google Api key which can be accquired here: https://console.cloud.google.com/ 

Also it needs a client_secrect.json file.

- API_KEY is the api key
- CLIENT_SECRET is the path to the "client_secrets.json" file

```python
# Youtube Api's credentials
api_credentials = {'api_key': 'YOUR_API_KEY',	
           'client_secret': "client_secrets.json"}

# Create YouTube
youtube = graphipy.get_youtube(api_credentials)
```

# Searching for videos with topic
## fetch_videos_by_topic(graph, topic, maxResult=25):
This method grabs videos about a certain topic and their corresponding creator (channel)
and add to the graph

Required Parameters:
- graph: The graphify graph we are passing in
- topic: The keyword for query

Optional Parameters:
- param maxResult: Max result for query, default 25

Type of Nodes Returning:
- Channel Node (Represents User)
- Video Node

Type of Edges Returning:
- Channel-Create-Video Edge 

```python
# Create a graph and naming the graph type inside graphipy first and then modify the graph
graph = graphipy.create_graph("pandas")
youtube.fetch_videos_by_topic(graph, "usc")
nodes = graph.get_nodes()
df = nodes["video"]
df.head(10)
```

```python
df = nodes["channel"]
df.head(10)
```

```python
edges = graph.get_edges()
df = edges["createvideo"]
df.head(25)
```

# Searching for Channels with topic
## fetch_channels_by_topic(self, graph, topic, maxResult=25):
This method fetches channels with a given topic 
returning a disconnected graph (which contains different disconnected channel nodes)

Required Parameters:
- graph: The graphify graph we are passing in
- topic: The keyword for query

Optional Parameters:
- param maxResult: Max result for query, default 25

Type of Nodes Returning:
- Channel Node

Type of Edges Returning:
- None

```python
graph_channel = graphipy.create_graph("pandas")
youtube.fetch_channels_by_topic(graph_channel,"usc")
channel_nodes = graph.get_nodes()
df = channel_nodes["channel"]
df.head(15)
```

# Searching for Playlists with topic
## fetch_playlists_by_topic(self, graph, topic, maxResult=25):
This method fetches playlists with a given topic and all their creators

Required Parameters:
- graph: The graphify graph we are passing in
- topic: The keyword for query

Optional Parameters:
- param maxResult: Max result for query, default 25

Type of Nodes Returning:
- Channel Node
- Playlist Node

Type of Edges Returning:
- Channel-Create_Playlist edge

```python
graph = graphipy.create_graph("pandas")

youtube.fetch_playlists_by_topic(graph, "usc")
nodes = graph.get_nodes()
df = nodes["playlist"]
df.head(20)
```

# Search Comment for a Video using Video id
## fetch_video_by_id_with_comments(self, graph, video_id)
This method fetches a video with its comments and its creator (channel node) and also all the creator (channel node) for the comments (people who comment the comment)

Required Parameters:
- graph: The graphify graph we are passing in
- topic: The keyword for query

Optional Parameters:
- param maxResult: Max result for query, default 25

Type of Nodes Returning:
- Channel Node
- Video Node
- Comment Node

Type of Edges Returning:
- Channel-Comment_Com edge

        :param graph: The graph we are passing in
        :param video_id: The id of the video
        :return: The resulting graph

```python
info_of_one_video_graph = graphipy.create_graph("pandas")
# let's search for "USC vs. Colorado | FOX COLLEGE FOOTBALL HIGHLI..." which has id of DAV2sWfiZN4
youtube.fetch_video_by_id_with_comments(info_of_one_video_graph, "DAV2sWfiZN4")
nodes = info_of_one_video_graph.get_nodes()
df = nodes["comment"]
df.head(20)
```

# Visualization with NetworkX

```python
from matplotlib import colors as mcolors
import matplotlib.pyplot as plt

youtube.fetch_video_by_id_with_comments(info_of_one_video_graph, "DAV2sWfiZN4")
nodes = info_of_one_video_graph.get_nodes()
edges = info_of_one_video_graph.get_edges()
```

```python
# Create the networkx exporter object
exporter = graphipy.get_nx_exporter()
nx_graph = exporter.create_from_pd(nodes_df=nodes, edges_df=edges)
```

```python
# Draw the graph using graphipy
color_set = set(mcolors.CSS4_COLORS)
options = {
    "node_label": "Label",
    "colorful_edges": True,
    "color_set": color_set
}
legend = exporter.draw_random(nx_graph, options=options, legend=plt)
```
